## Respuesta al Escalón

La respuesta de un SLTI a un [Heaviside signal](../continuous_signals/standard_signals.ipynb#Heaviside-Signal) $\epsilon(t)$ a la entrada es conocida como [*respuesta al escalón*](https://en.wikipedia.org/wiki/Step_response). Se define como

\begin{equation}
h_\epsilon(t) = \mathcal{H} \{ \epsilon(t) \}
\end{equation}

La respuesta al escalón caracteriza las propiedades de un sistema cuando la señal de entrada se 'enciende' en $t=0$. Puede estar relacionado con la respuesta de impulso $h(t)$ del sistema por

\begin{equation}
h_\epsilon(t) = \epsilon(t) * h(t) = \int_{-\infty}^{t} h(\tau) \; d\tau
\end{equation}

La anterior igualdad se puede deducir, por ejemplo, de una [interpretación gráfica de la convolución](convolution.ipynb#Graphical-Interpretation). Como la operación inversa a la integral anterior es la derivada, esto implica que la respuesta al impulso es la derivada de la respuesta al escalón.

\begin{equation}
h(t) = \frac{d h_\epsilon(t)}{dt}
\end{equation}

Al usar este resultado junto con las propiedades de la convolución para derivadas, la señal de salida $y(t)=\mathcal{H}\{x(t)\}$ en términos de la respuesta al escalón se tiene que

\begin{equation}
y(t) = x(t) * \frac{d h_\epsilon(t)}{dt} = \frac{d x(t)}{dt} * h_\epsilon(t)
\end{equation}

La señal de salida de un SLTI está dada por la convolución entre la derivada de la señal de entrada con la respuesta al escalón.

La respuesta al escalón es una alternativa para medir las propiedades de un sistema LTI, ya que en la práctica no se puede realizar un impulso de Dirac. Juega un papel importante en la teoría de [sistemas de control](https://en.wikipedia.org/wiki/Control_system).
