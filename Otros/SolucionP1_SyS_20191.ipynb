## Pregunta 1

Clasificar las señales de acuerdo a las siguientes categorias: i) periódica, aperiódica, o cuasi-periódica, ii) energía o potencia. a) $x[n]=A \cos[n\pi]u[n-n_o],$ donde $A\in \mathbb{R}^+,$ $n\in\{0\pm1,\pm2,\dots,\pm N\},$ y $0<n_o<N.$ b) $x(t) = Bte^{-at}(u(t)-u(t-t_o)),$ con $t_o>0$ y $B,a \in \mathbb{R}$


## Solución

a) $x[n]=A \cos[n\pi]u[n-n_o].$ Dado que el escalón se activa en $n_o$, la estructura periódica del coseno se pierde para $t<0,$ por ende $x[n]$ es **aperiódica**. Ahora, se cálcula la energía  de la señal en función de $N$ como:

\begin{align}
E(N)=& \sum^N_{n=-N}{\left|A \cos[n\pi]u[n-n_o]\right|^2}=\sum^N_{n=n_o}{\left|A \cos[n\pi]\right|^2}\\
E(N)=& \sum^N_{n=n_o}{\frac{A^2}{2} +\frac{A^2}{2}\cos[2n\pi]}= \sum^N_{n=n_o}A^2 = A^2(N-n_o+1)\\
\end{align}
dado que $\cos[2n\pi]=1.$ Evaluando el límite $N\rightarrow \infty$ (derivando numerador y denomirador con respecto a N) claramente la energía diverge. Por ende, $x[n]$ **no es de energía**. Ahora, para calcular la potencia en función de N y aplicando el límite  nuevamente se tiene que:

\begin{align}
P(N)=& \frac{1}{2N+1}E(N) = \frac{A^2}{2N+1}(N-n_o+1)\\
\lim_{N\rightarrow \infty}P(N)=& \lim_{N\rightarrow \infty} \frac{A^2}{2}= \frac{A^2}{2}.\\
\end{align}

Por consiguiente, $x[n]$ es **señal de potencia** siempre y cuando $0<A<\infty.$

b) $x(t) = Bte^{-at}(u(t)-u(t-t_o)).$  Para el cálculo de la energía en función de $t_o$ (variable que controla el trozo de señal a considerar distanta de cero), se tiene que:

\begin{align}
E(t_o)=& \int^{T/2}_{-T/2}{\left|Bte^{-at}(u(t)-u(t-t_o))\right|^2}dt = \int^{t_o}_{0}{\left|Bte^{-at}\right|^2}dt \\
E(t_o)=& \int^{t_o}_{0}{B^2 t^2 e^{-2at}} dt = -\frac{\left(2B^2a^2t^2+2B^2at+B^2\right)e^{-2at}}{4a^3}\Biggr|_{0}^{t_o} \\
E(t_o)=& \frac{B^2}{4a^3}-\frac{\left(2B^2a^2t_o^2+2B^2at_o+B^2\right)e^{-2at_o}}{4a^3} \\
\end{align}

La anterior expresión se obtiene mediante integración por partes. Evaluando el límite para $t_o\rightarrow \infty$:

\begin{equation}
\lim_{t_o\rightarrow \infty}E(t_o)= \lim_{t_o\rightarrow \infty} \frac{B^2}{4a^3}-\frac{\left(2B^2a^2t_o^2+2B^2at_o+B^2\right)e^{-2at_o}}{4a^3}= \frac{B^2}{4a^3}.
\end{equation}

El límite anterior se obtiene derivando dos veces el numerador y denominador de la expresión $E(t_o)$ y se restringe para $a>0.$ Por ende, $x(t)$ es **señal de energía**. Como la energía es finita para $\lim_{t_o\rightarrow \infty}E(t_o)$, entonces $\lim_{t_o\rightarrow \infty}P(t_o)=\frac{1}{t_o}E(t_o) = 0.$ Por consiguiente, $x(t)$  **no es señal de potencia**.



```python
#Simulación pregunta 1
# %% definir librerias
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#grafica señal a)
N = 10
n0 = 2
A  = 3
n = np.linspace(-N,N,2*N+1)
u = n >= n0 
x = A*np.cos(n*np.pi)*u

plt.stem(n,x,label="$\cos[n\pi](u[n-n_o])$")
plt.grid()
plt.legend(loc='upper left')
plt.xlim(-N-1,N+1)
plt.xlabel("n")
plt.ylabel("$x[n]$")
#plt.ylim(-0.25,6.5)

print("La energía para N=%d es E(N)=%.2f" % (N,np.sum(x**2)))
print("La energía para N=%d según formula E(N) = A^2(N-no+1) =%.2f\n" % (N,(A**2)*(N-n0+1)))

print("La potencia para N=%d es P(N)=%.2f" % (N,np.mean(x**2)))
print("La potencia para N=%d según formula P(N) = A^2(N-no+1)/(2N+1) =%.2f" % (N,(A**2)*(N-n0+1)/(2*N+1)))
```

## Pregunta 2

Hallar la salida $y[n]$ del sistema lineal e invariante con el tiempo $\tau\{\cdot\}$ con respuesta al escalón $s[n]=\{1,-2,0^\xi,1,-1\}$,  ante la entrada $x[n]=\{-1,3^\xi,2,-1\}$, donde $n\in\{0,\pm1,\pm2,\dots,\pm N\}$ y $n=0$ para $x[n]^\xi.$



## Solución

Para resolver el sistema ante cualquier entrada, se necesita la respuesta al impulso. En este sentido, la respuesta al impulso $h[n]$ se obtiene desde la respuesta al escalón $s[n]$ según: 

\begin{equation}
h[n] = s[n]-s[n-1]=\{1,-3,2^\xi,1,-2,1\}.
\end{equation}

Asi pues:

\begin{equation}
y[n]= \tau\{x[n]\}= x[n]*h[n]=\sum_{k=-5}^{5}x[k]h[n-k]; \quad \forall n\in\{0,\pm 1, \pm 2, \pm 3, \pm 4, \pm 5\}.
\end{equation}

Por ende:

\begin{equation}
y[n] = \{-1,  6, -9, -2^{\xi}, 12, -7, -2,  4,-1\}
\end{equation}



```python
#simulación pregunta 2
s = [0,1,-2,0,1,-1,0]
x = [0,0,-1,3,2,-1,0]
N = int(len(s)/2)
n = np.linspace(-N,N,2*N+1)

h = np.append(0,np.diff(s)) 

plt.stem(n, s, label=r'$s[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$s[n]$')
plt.grid()
plt.show()

plt.stem(n, h, label=r'$h[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$h[n]$')
plt.grid()
plt.show()

plt.stem(n, x, label=r'$x[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$x[n]$')
plt.grid()
plt.show()


y = np.convolve(x,h)
ny = np.linspace(-2*N,2*N,4*N+1)
plt.stem(ny, y, label=r'$y[n]$')
plt.xlim([-2*N-1,2*N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$y[n]$')
plt.grid()
plt.show()

print("h[n]=",h)
print("y[n]=",y)

```

## Pregunta 3

Se tiene un microprocesador de 2 bits con entrada análoga entre 0 y 5 [v] y frecuencia de muestreo 80 [Hz]. Se presente digitalizar la señal $x(t)=9\cos(20\pi t) - 3\cos(100\pi t)$.  Realice el diseño del proceso de digitalización, incluyendo bosquejos de las graficas de discretización y cuantización, y encuentre una expresión para $x[n]$ discreta y digital. En caso de que el sistema de digitalización no sea apropiado, proponga los cambios necesarios.

## Solución

Se requiere acomodar la señal entre 0 y 5 [v]. Por consiguiente, se diseña la etapa de cero y pendiente de la forma: $y(t)=mx(t)+b$, tal que $y(t)\in[0,5].$ La pendiente $m$ se puede calcular desde los valores mínimo y máximo de x(t), donde $x_{max}(t)=10.45$ y $x_{min}(t)=-10.45$. Por ende:

\begin{equation}
m = \frac{5-0}{x_{max}-x_{min}}
\end{equation}

y

\begin{equation}
b = y - m x_{xmax} = 5 - m x_{max}
\end{equation}

Ahora, discretizando la señal

\begin{equation}
y[n/Fs] = m(9\cos[20\pi n/80]-3\cos[100\pi n/80])+b = m(9\cos[\pi n/4]-3\cos[5\pi n/4])+b
\end{equation}
donde:

\begin{equation}
\cos[5\pi n/4] = \cos[(5\pi/4-2\pi)n] = \cos[-3n\pi/4] = \cos[3n\pi/4] = -\cos[n\pi/4],
\end{equation}
por consiguiente:

\begin{equation}
y[n/Fs] = m(9\cos[\pi n/4]+3\cos[\pi n/4])+b = m 12\cos[\pi n/4]+b
\end{equation}

Dado que $Fs < 2 F_{max}$, la información en el sistema de discretización es alterada. Se requiere que $Fs\geq 100$[Hz].

Con respecto a la etapa de cuantización, se obtienen 4 posibles estados ($2^2=4$), con tamaño de paso en voltaje igual a $\Delta v = 5/4 = 1.25 [v]$

\begin{array}{c|c} \hline
x[n] & x_c[n] & Binario\\ \hline
[0.00,1.25) & 0.00 & 0 0\\ \hline
[1.25,2.50) & 1.25 & 0 1\\ \hline
[2.50,3.75) & 2.50 & 1 0\\ \hline
[3.75,5.00] & 3.75 &1 1\\ \hline
\end{array}


```python
#simulación pregunta 3
import numpy as np
import matplotlib.pyplot as plt


fc1 = 10
fc2 = 50
fss = 10*fc2 #frecuencia de muestreo para emular tiempo continuo
fs = 80 # frecuencia muestreo ejercicio
Tplot = 2/fc1 #tiempo a graficar
t = np.arange(0,Tplot,1/fss)

x1 = 9*np.cos(2*np.pi*fc1*t)
x2 = -3*np.cos(2*np.pi*fc2*t)
x = x1 + x2

xmin = min(x)
xmax = max(x)
ymin = 0
ymax = 5
m = (ymax - ymin)/(xmax-xmin)
b = ymax - m*xmax
y = m*x+b

plt.plot(t,x,label="x(t)")
plt.plot(t,y,label="y(t)")
plt.legend()
plt.xlabel('t')
plt.grid()
plt.title("Cero y pendiente")
plt.show()

tm = np.arange(0,Tplot,1/fs)
x1m = 9*np.cos(2*np.pi*fc1*tm)
x2m = -3*np.cos(2*np.pi*fc2*tm)
xm = x1m + x2m
ym = m*xm+b

plt.plot(t,y,color='r',label="y(t)")
plt.stem(tm,ym,label="y[nTs]")
plt.legend()
plt.xlabel('t')
plt.grid()
plt.title("cero y pendiente + discretizacion señal completa fs = 80 Hz")
plt.show()

plt.plot(t,x1,color='r',label='$x(t)=9cos(20\pi t)$')
plt.stem(tm,x1m,label="$x[n/80]=9cos[20\pi n/80]$")
plt.legend(loc='best')
plt.grid()
plt.title("Discretizacion fs = 80 Hz $x(t)=9cos(20\pi t)$")
plt.show()

plt.plot(t,x2,color='r',label='$x(t)=-3cos(100\pi t)$')
plt.stem(tm,x2m,label="$x[n/80]=-3cos[100\pi n/80]$")
plt.legend(loc='best')
plt.grid()
plt.title("Discretizacion fs = 80 Hz $x(t)=-3cos(100\pi t)$")
plt.show()


# cero y pendiente mas discretizacion mas cuantizacion
yc = np.zeros(len(y))

for i in range(len(y)):
    if y[i] < 1.25:
       yc[i] = 0
    elif (y[i] >= 1.25) and (y[i] <= 2.5):
        yc[i] = 1.25
    elif (y[i] >= 2.5) and (y[i] <= 3.75):
        yc[i] = 2.5
    else:
        yc[i] = 3.75
    
plt.plot(t,y,color='r',label="y(t)")
plt.plot(t,yc,label="$y_c[nTs]$",linewidth = 3)
plt.legend()
plt.xlabel('t')
plt.grid()
plt.title("cero y pendiente + discretizacion (fs=80Hz) + cuantizacion (2 bits)")
plt.show()


```
